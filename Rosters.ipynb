```python
import json
import pandas as pd
fulldata = pd.read_csv("data/2017-18_playerBoxScore.csv", encoding="utf-8")
```

```python
teams_players = fulldata.groupby(["teamAbbr","playDispNm"]).size().reset_index()[["teamAbbr","playDispNm"]]
roster_dict = teams_players.groupby("teamAbbr")["playDispNm"].apply(lambda x: x.tolist()).to_dict()
roster_dict
```

```python
with open("static/js/roster1718.js","w",encoding="utf8") as x:
    x.write("rosterdata = [")
    json.dump(roster_dict, x, ensure_ascii=False)
    x.write("]")
```
