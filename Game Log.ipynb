```python
import json
import pandas as pd
raw_data = pd.read_csv("data/2017-18_playerBoxScore.csv", encoding="utf-8")
```

```python
all_data = raw_data[["playDispNm","playPos","gmDate","playMin","playStat","playPTS","playTRB","playAST","playSTL","playBLK","playTO","playPF","playFGM","playFGA","play3PM","play3PA","playFTM","playFTA","playORB","teamAbbr","opptAbbr","teamLoc","teamRslt"]]
all_data.head()
```

```python
team_list = sorted(all_data["teamAbbr"].unique().tolist())
```

```python
teamdata = {}
teamplayers = {}
player_list = []
teamdata["ALL"] = {}
for team in team_list:
    teamdata[team] = {}
    
    teamdataframe = all_data[all_data["teamAbbr"]==team]
    players = sorted(all_data[all_data["teamAbbr"]==team].playDispNm.unique().tolist())
    for player in players:
        teamdata[team][player] = teamdataframe[teamdataframe["playDispNm"]==player]
        teamdata["ALL"][player] = all_data[all_data["playDispNm"]==player]
        
    player_list.extend(players)
```

```python
clean_list = {}
for name in player_list:
    inputname=name
    name=name.replace("Luc Richard Mbah a Moute","LucRichardMbahAMoute");
    name=name.replace("Nazareth","Naz");
    name=name.replace("Raulzinho","Raul");
    name=name.replace("James Web","James Webb");
    for x in "ÀÁÂÃÄÅĀ":
        name=name.replace(x,"A");
    for x in "àáâãäåā":
        name=name.replace(x,"a");
    for x in "ëèéê":
        name=name.replace(x,"e");
    for x in "ôóøŏöò":
        name=name.replace(x,"o");
    for x in "úùüū":
        name=name.replace(x,"u");
        
    name=name.replace("ģ","g");
    name=name.replace("č","c");
    name=name.replace("ć","c");
    name=name.replace("š","s");
    name=name.replace("ņ","n");
    name=name.replace("ñ","n");
    name=name.replace("í","i");
    name=name.replace("Ž","Z");
    name=name.replace("ž","z");
    name=name.replace("ý","y");
    name=name.replace("'","");
    name=name.replace("’","");
    name=name.replace(".","");
    name=name.replace("Jr","");
    name=name.replace("III","");
    name=name.replace("II","");
    name=name.replace("-","");
    name=name.replace(",","");
    name=name.replace(" ","");
    
    clean_list[inputname] = name
```
